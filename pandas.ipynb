```python
import pandas as pd

# 1. 데이터 생성
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 40, 28],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],
    "Salary": [50000, 60000, 70000, 80000, 55000]
}

df = pd.DataFrame(data)

# 2. 데이터 확인
print("=== 데이터프레임 확인 ===")
print(df)          # 전체 데이터프레임 출력
print(df.head(3))  # 상위 3개 행 출력
print(df.info())   # 데이터 타입과 결측값 정보 확인
print(df.describe())  # 기본 통계 정보 출력

# 3. 특정 열 선택
print("\n=== 특정 열 선택 ===")
print(df["Name"])  # 'Name' 열만 출력
print(df[["Name", "Salary"]])  # 여러 개의 열 선택

# 4. 행 선택 (인덱싱 및 슬라이싱)
print("\n=== 행 선택 ===")
print(df.loc[1])  # 인덱스 1번 행 선택 (라벨 기반)
print(df.iloc[2])  # 인덱스 2번 행 선택 (숫자 기반)
print(df[1:4])  # 인덱스 1~3번 행 선택

# 5. 조건 필터링
print("\n=== 조건 필터링 ===")
print(df[df["Age"] >= 30])  # Age가 30 이상인 행 선택
print(df[(df["Age"] >= 30) & (df["Salary"] > 60000)])  # 여러 조건 적용

# 6. 새로운 열 추가 및 기존 열 수정
df["Bonus"] = df["Salary"] * 0.1  # Salary의 10%를 Bonus로 추가
df["Age Category"] = df["Age"].apply(lambda x: "Young" if x < 30 else "Old")  # 나이 기준으로 카테고리 나누기
print("\n=== 새로운 열 추가 ===")
print(df)

# 7. 데이터 정렬
print("\n=== 데이터 정렬 ===")
print(df.sort_values(by="Salary", ascending=False))  # Salary 기준 내림차순 정렬
print(df.sort_values(by=["Age", "Salary"], ascending=[True, False]))  # Age 오름차순, Salary 내림차순 정렬

# 8. 그룹화 및 집계
print("\n=== 그룹화 및 집계 ===")
grouped = df.groupby("City")["Salary"].mean()  # 도시별 평균 연봉 계산
print(grouped)

# 9. 결측값 처리
df.loc[2, "Salary"] = None  # 결측값 추가
print("\n=== 결측값 처리 ===")
print(df.isnull().sum())  # 결측값 개수 확인
df["Salary"].fillna(df["Salary"].mean(), inplace=True)  # 결측값 평균으로 대체
print(df)

# 10. 중복값 제거
df.loc[5] = df.loc[3]  # 중복 행 추가
print("\n=== 중복값 제거 ===")
print(df.duplicated())  # 중복 여부 확인
df.drop_duplicates(inplace=True)  # 중복 제거
print(df)

# 11. CSV 파일 저장 및 불러오기
df.to_csv("output.csv", index=False)  # CSV로 저장
df_loaded = pd.read_csv("output.csv")  # 저장한 파일 불러오기
print("\n=== CSV 파일 불러오기 ===")
print(df_loaded)
```
